```python
import pandas as pd

datasets = ['amazon', 'celeba', 'gaussian', 'moons']
# Path to your CSV file
data = datasets[3]
csv_file_path = f'./history/baseline_metrics_{data}.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

df = df[df['mode']=='ind']
# Get the column names
columns = list(df.columns)

# Move the second-to-last column to the 9th position (index 8)
col_to_move = columns[-2]
columns.insert(8, columns.pop(columns.index(col_to_move)))

# Reorder the DataFrame
df = df[columns]

print(df.head())
```

```python
# Select all columns except the first four

columns_to_analyze = df.columns[4:-4]

# Group by 'mode'
grouped = df.groupby('amazon_cat')

# Calculate mean and standard deviation for each group
mean_df = grouped[columns_to_analyze].mean()
std_df = grouped[columns_to_analyze].std()

# Create a new DataFrame to store the formatted mean±std values
formatted_df = mean_df.copy()

# Format the mean and standard deviation to two decimal places and combine them
for col in columns_to_analyze:
    formatted_df[col] = mean_df[col].map('{:.2f}'.format) + '±' + std_df[col].map('{:.2f}'.format)

# Print the formatted DataFrame
print(f'dataset name:  {data}')
print("Formatted Mean±Std values:")
print(formatted_df)
```
